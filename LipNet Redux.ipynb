## Exploring the dataset and Data Preprocessing and defining important functions

The file bbaf2n.align contains:

    0 23750 sil
    23750 29500 bin
    29500 34000 blue
    34000 35500 at
    35500 41000 f
    41000 47250 two
    47250 53000 now
    53000 74500 sil

sil at the beginning and end denotes silence and the other words represent the words spoken by the speaker.

For each .align file, data contains a cooresponding video file that shows the lip movement while making those sounds

```python
def load_video(path:str) -> List[float]:
    """Outputs: A list of floats which represent our video"""
    cap = cv2.VideoCapture(path)
    frames=[]
    for _ in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))): #Iterate through all frames of the video
        ret, frame = cap.read()
        frame = tf.image.rgb_to_grayscale(frame)
        frames.append(frame[180:240, 80:220, :]) #Isolating the mouth region, TODO: Make this automatic using face detector
    cap.release()
    
    #Standardizing the input
    mean = tf.math.reduce_mean(frames)
    std = tf.math.reduce_std(tf.cast(frames, tf.float32))
    return tf.cast((frames-mean), tf.float32)/std
```

```python
vocab = [x for x in "abcdefghijklmnopqrstuvwxyz'?!123456789 "]
```

```python
char_to_num = tf.keras.layers.StringLookup(vocabulary=vocab, oov_token="") #oov_token means if they have never seen something before it will just be blank
```

```python
num_to_char = tf.keras.layers.StringLookup(vocabulary=char_to_num.get_vocabulary(), oov_token="", invert=True)
```

The above two functions are useful for converting char to num and num to char, which are useful to convert our strings into number, part of data preprocessing

```python
char_to_num(['a','b','z'])
```

This is useful for calculating the loss function! since our model will calculate this, essentially these will act as lookup functions

```python
#FUNCTION TO LOAD UP ALIGNMENTS
def load_alignments(path:str) -> List[str]:
    with open(path, 'r') as f:
        lines = f.readlines()
    tokens = []
    for line in lines:
        line = line.split()
        if line[2] != 'sil':
            tokens = [*tokens, ' ', line[2]]
    return char_to_num(tf.reshape(tf.strings.unicode_split(tokens, input_encoding='UTF-8'), (-1))) [1:]
```

The above function opens the input alignment file and returns a 1D Tensor, which contains the numbers cooresponding to each letter that is said in the video

```python
#Function to load one input file (both video and alignment file at once)
def load_data(path: str):
    path = bytes.decode(path.numpy())
    file_name = path.split('\\')[-1].split('.')[0]
    '''NOTE: If you're using a Linux Machine
        Change: file_name = path.split('/')[-1].split('.')[0]
    '''
    video_path = os.path.join('data', 's1', f'{file_name}.mpg')
    align_file_path = os.path.join('data','alignments', 's1', f'{file_name}.align')
    frames = load_video(video_path)
    alignments = load_alignments(align_file_path)
    
    return frames, alignments
```

```python
def mappable_function(path:str) -> List[str]:
    return tf.py_function(load_data, [path], (tf.float32, tf.int64))
#Required for tensorflow data pipeline
```

### Creating the Dataset and Understanding the Dataset More

```python
data = tf.data.Dataset.list_files('./data/s1/*.mpg')
data = data.shuffle(500)
data = data.map(mappable_function) #This will return frames and alignments
data = data.padded_batch(2, padded_shapes=([75, None, None, None], [40])) #Padding the data together since the number of frames in each video are variable
#Each data point now has two videos and atleast 75 frames and 40 alignments
```

## Designing the Deep Neural Network

We want a neural network that takes in val[0] (i.e. frames) and outputs string(val[1]) i.e. the string that the person is saying in those two videos combined

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv3D, LSTM, Dense, Dropout, Bidirectional, MaxPool3D, Activation, Reshape, SpatialDropout3D, BatchNormalization, TimeDistributed, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, LearningRateScheduler
```

```python
model = Sequential()
model.add(Conv3D(128, 3, input_shape=(75,60,140,1), padding='same'))
model.add(Activation('relu'))
model.add(MaxPool3D((1,2,2)))

model.add(Conv3D(256, 3, padding='same'))
model.add(Activation('relu'))
model.add(MaxPool3D((1,2,2)))

model.add(Conv3D(75, 3, padding='same'))
model.add(Activation('relu'))
model.add(MaxPool3D((1,2,2)))

model.add(TimeDistributed(Flatten()))

model.add(Bidirectional(LSTM(128, kernel_initializer='Orthogonal', return_sequences=True)))
model.add(Dropout(.5))

model.add(Bidirectional(LSTM(128, kernel_initializer='Orthogonal', return_sequences=True)))
model.add(Dropout(.5))

model.add(Dense(char_to_num.vocabulary_size()+1, kernel_initializer='he_normal', activation='softmax'))
```

```python
model.summary()
```

For each of the 75 Frames, we output one of 41 possible numbers/characters
